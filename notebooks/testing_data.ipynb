```python
import pandas as pd

planets_df = pd.read_csv('../dataset/planets.csv')

print(planets_df.head(10))
```

```python
sorted_platets_mass = planets_df.sort_values('mass_10^24kg', ascending=False)
print(sorted_platets_mass)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data=planets_df, x='mass_10^24kg', bins=20, kde=True)
plt.xlabel('Mass (10^24 kg)')
plt.ylabel('Count')
plt.title('Distribution of Planet Masses')
plt.show()
```

```python
planet_temperatures = planets_df[['planet', 'temperature_avg_c' ]]
planet_temperatures = planet_temperatures.sort_values('temperature_avg_c', ascending=True)
print(planet_temperatures.head(10))
```

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test for normality on the temperature data
stat, p_value = shapiro(planets_df['temperature_avg_c'])
print(f'Statistics={stat}, p-value={p_value}')
if p_value > 0.05:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```
